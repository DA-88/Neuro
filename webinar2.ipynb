```python
#!c1.8
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.utils import to_categorical
from PIL import Image
import numpy as np
import os
import random
```

#### Готовим списки

```python
#!c1.8
i = 0
x_train = []
y_train = []
x_test  = []
y_test  = []

while i <= 9:
    for root, dirs, files in os.walk(f"train_captcha/{i}"):
      for file in files:
        y_train.append(i)
        with Image.open(os.path.join(root, file)) as a:     
            x_train.append(np.array(a))    

    for root, dirs, files in os.walk(f"test_captcha/{i}"):
      for file in files:
        y_test.append(i)
        with Image.open(os.path.join(root, file)) as a:     
            x_test.append(np.array(a))
             
    i += 1
```

```python
#!c1.8
t = list(zip(x_train, y_train))
random.shuffle(t)
x_train1=[]
y_train1=[]
for tt in t:
    x_train1.append(tt[0])
    y_train1.append(tt[1])
```

```python
#!c1.8
model = keras.models.load_model('captcha.h5')
```

```python
# #!c1.8
model = keras.Sequential([
    keras.layers.Flatten(input_shape=(44, 24)),
    keras.layers.Dense(528, activation='tanh'),
    keras.layers.Dense(264, activation='tanh'),
    keras.layers.Dense(132, activation='tanh'),
    keras.layers.Dense(66, activation='tanh'),
    keras.layers.Dense(33, activation='tanh'),
    keras.layers.Dense(10,  activation='softmax')
])

model.compile(optimizer = 'adam',
              loss='categorical_crossentropy',
              metrics = ['accuracy'])
```

```python
# #!c1.8
model.fit(np.array(x_train1), to_categorical(y_train1), epochs=128, batch_size=42510, use_multiprocessing=True, shuffle=True, verbose=1)
```

```python
#!c1.8
model.evaluate(np.array(x_test), to_categorical(y_test))
```
